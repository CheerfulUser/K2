# Spot-check Everest Validation Summaries for KEGS

This notebook does more spot-checking of KEGS target lightcurves with the "data validation summary" (DVS) feature in EVEREST.

```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
```

```python
import astropy
import os
import pandas as pd
import seaborn as sns
from astropy.utils.console import ProgressBar
import everest
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
everest_path = '../../everest/everest/missions/k2/tables/'
df_kegs = pd.read_csv('../metadata/KEGS_TPF_metadata.csv')
c05_everest = pd.read_csv(everest_path + 'c05.stars', names=['EPIC_ID', 'KepMag', 'Channel', 'col4'])
kegs_everest_c05 = pd.merge(df_kegs, c05_everest, how='inner', left_on='KEPLERID', right_on='EPIC_ID')
```

```python
ke_list = kegs_everest_c05['KEPLERID'].values
```

```python
i = 0
```

```python
i
```

Decent examples to try to replicate: 12 

```python
i+=1
star = everest.Everest(ke_list[i])
star.dvs()
```

```python
for i in range(len(ke_list)):
    star = everest.Everest(ke_list[i])
    star.dvs()
```
